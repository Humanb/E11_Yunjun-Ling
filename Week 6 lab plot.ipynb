## Indoor Vs Outdoor data

```python
plt.hist(indoor_data["temperature"], bins=30, density = True)
plt.ylabel('Temperature')
plt.xlabel('Time')
plt.title("Temperature vs Time")
x = np.linspace(np.min(indoor_data["temperature"]), np.max(indoor_data["temperature"]), 100)
mean = np.mean(indoor_data["temperature"])
std = np.std(indoor_data["temperature"])
y = norm.pdf(x, mean, std)
plt.plot(x,y, label = "Indoor Norm", color ='k')
plt.hist(outdoor_data["temperature"], bins = 30, density = True)
plt.ylabel('Temperature')
plt.xlabel('Time')
plt.title("Temperature vs Time")
x = np.linspace(np.min(outdoor_data["temperature"]), np.max(outdoor_data["temperature"]), 100)
mean = np.mean(outdoor_data["temperature"])
std = np.std(outdoor_data["temperature"])
y = norm.pdf(x, mean, std)
plt.plot(x,y, label = "Outdoor Norm", color ='cyan')
plt.legend()
plt.show()
print("Indoor temperature mean: ", np.mean(indoor_data["temperature"]))
print("Outdoor temperature mean: ", np.mean(outdoor_data["temperature"]))
print("Indoor temperature std: ", np.std(indoor_data["temperature"]))
print("Outdoor temperature std: ", np.std(outdoor_data["temperature"]))
```

```python
plt.hist(indoor_data["pressure"], bins=50, density = True)
plt.ylabel('Pressure')
plt.xlabel('Time')
plt.title('Pressure vs Time')
x1 = np.linspace(np.min(indoor_data["pressure"]), np.max(indoor_data["pressure"]), 100)
mean = np.mean(indoor_data["pressure"])
std = np.std(indoor_data["pressure"])
y1 = norm.pdf(x1, mean, std)
plt.plot(x1,y1, label = "Indoor pressure norm", color ='k')
plt.hist(outdoor_data["pressure"], bins=50, density = True)
plt.ylabel('Pressure')
plt.xlabel('Time')
plt.title('Pressure vs Time')
x1 = np.linspace(np.min(outdoor_data["pressure"]), np.max(outdoor_data["pressure"]), 100)
mean = np.mean(outdoor_data["pressure"])
std = np.std(outdoor_data["pressure"])
y1 = norm.pdf(x1, mean, std)
plt.plot(x1,y1, label = "Outdoor pressure norm", color ='b')
plt.legend()
plt.show()
print("Indoor pressure mean: ", np.mean(indoor_data["pressure"]))
print("Outdoor pressure mean: ", np.mean(outdoor_data["pressure"]))
print("Indoor pressure std: ", np.std(indoor_data["pressure"]))
print("Outdoor pressure std: ", np.std(outdoor_data["pressure"]))
```
